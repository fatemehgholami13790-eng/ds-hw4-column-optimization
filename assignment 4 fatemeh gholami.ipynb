```python
from collections import defaultdict

def solve():
    # Read the number of blocks
    N = int(input())
    
    # Initialize a hash table (dictionary) to group blocks by base dimensions (x, y)
    blocks_by_base = defaultdict(list)
    
    # Reading input and filling the hash table
    for _ in range(N):
        x, y, z = map(int, input().split())
        # Sort (x, y, z) to ensure x >= y >= z
        x, y = max(x, y), min(x, y)
        blocks_by_base[(x, y)].append(z)
    
    # Initialize a variable to store the result
    max_stack_height = 0
    
    # Now, for each base (x, y), we need to calculate the maximum stack height
    for (x, y), heights in blocks_by_base.items():
        # Sort heights in descending order
        heights.sort(reverse=True)
        
        # Calculate the stackable height for each block (sum of z values for stacking)
        current_stack_height = 0
        for i in range(len(heights)):
            current_stack_height += heights[i]
            # We are interested in the maximum stack height that respects the base dimensions
            max_stack_height = max(max_stack_height, min(y, current_stack_height))
    
    # Print the result
    print(max_stack_height)

# Example usage
if __name__ == "__main__":
    solve()
```

```python
class HashTable:
    def __init__(self, size):
        # ایجاد جدول هش با اندازه مشخص
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, x, y):
        # تابع هش برای محاسبه اندیس کلید
        return (x * 31 + y) % self.size

    def insert(self, x, y, z, index):
        # افزودن بلوک با ابعاد (x, y) و ارتفاع z و ایندکس بلوک
        index_ = self._hash(x, y)
        self.table[index_].append((x, y, z, index))

    def search(self, x, y):
        # جستجو در جدول هش بر اساس (x, y)
        index_ = self._hash(x, y)
        return self.table[index_]


def solve():
    # خواندن تعداد بلوک‌ها
    N = int(input())
    
    # ساخت جدول هش با اندازه مناسب
    hash_table = HashTable(100)
    
    # خواندن ابعاد بلوک‌ها و افزودن به جدول هش
    blocks = []
    for i in range(N):
        x, y, z = map(int, input().split())
        x, y = max(x, y), min(x, y)
        blocks.append((x, y, z))
        hash_table.insert(x, y, z, i + 1)  # ایندکس‌ها 1-based هستن
    
    # متغیر برای ذخیره نتیجه
    max_diameter = 0
    result_blocks = []
    
    # بررسی هر بلوک و پیدا کردن جفت‌ها برای چیدمان
    for i in range(N):
        x1, y1, z1 = blocks[i]
        # جستجو در جدول هش برای پیدا کردن بلوک‌های با پایه مشابه
        group = hash_table.search(x1, y1)
        
        for x2, y2, z2, index2 in group:
            if index2 > i:  # فقط جفت‌های مختلف (غیرازدور) بررسی می‌شوند
                # محاسبه قطر قابل استفاده
                diameter = min(y1, z1 + z2)
                
                # مقایسه با حداکثر قطر قبلی
                if diameter > max_diameter:
                    max_diameter = diameter
                    result_blocks = [i + 1, index2]  # ایندکس‌ها 1-based هستن

    # چاپ کردن خروجی
    if result_blocks:
        print(len(result_blocks))
        print(" ".join(map(str, sorted(result_blocks))))
        print(max_diameter)
    else:
        print(1)
        print(1)
        print(blocks[0][2])  # فقط یک بلوک استفاده میشه


# اجرای کد
if __name__ == "__main__":
    solve()
```
